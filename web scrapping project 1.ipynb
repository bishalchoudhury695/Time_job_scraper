```python
import os
def find_jobs():
    # Ensure the 'posts' directory exists to prevent FileNotFoundError
    if not os.path.exists('posts'):
        os.makedirs('posts')  # Create the 'posts' directory if it doesn't exist
    
    # Fetch the HTML content from the job site
    html_text = requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&searchTextSrc=as&searchTextText=Python&txtKeywords=Python&txtLocation=').text
    
    # Parse the HTML content using BeautifulSoup with the lxml parser
    soup = BeautifulSoup(html_text, 'lxml')
    
    # Find all job listings on the page
    jobs = soup.find_all('li', class_='clearfix job-bx wht-shd-bx')
    
    # Define a skill you are unfamiliar with (this could be user input as well)
    unfamiliar_skill = "React"  # Example of an unfamiliar skill
    
    # Loop through each job found on the page
    for index, job in enumerate(jobs):
        # Find the published date to check how recent the job posting is
        published_date = job.find('span', class_= 'sim-posted').span.text
        
        # Only proceed if the job was posted recently (e.g., "few days ago")
        if 'few' in published_date:
            # Extract the company name and clean up any unwanted whitespace
            company_name = job.find('h3', class_='joblist-comp-name').text.replace('', '') 
            
            # Extract the skills required for the job and clean up whitespace
            skills = job.find('span', class_='srp-skills').text.replace('','') 
            
            # Get the URL for more information about the job
            more_info = job.header.h2.a['href']
            
            # Only save the job if the unfamiliar skill is NOT in the required skills
            if unfamiliar_skill not in skills:
                # Save the job details to a text file
                with open(f'posts/{index}.txt', 'w') as f:
                    # Write job details to the file
                    f.write(f"Company Name: {company_name.strip()}\n")  # Fixed typo: changed .Write() to .write()
                    f.write(f"Required Skills: {skills.strip()}\n")
                    f.write(f"More info: {more_info}\n")
                
                # Print confirmation that the file has been saved
                print(f'File saved: {index}')

if __name__ == '__main__':
    # Infinite loop to keep checking for jobs
    while True:
        find_jobs()  # Call the function to find jobs
        
        # Wait for 1 minute before checking again
        time_wait = 1
        print(f'Waiting {time_wait} minute(s)...')
        
        # Sleep for the specified number of minutes (converted to seconds)
        #time.sleep(time_wait * 60)
```
